```python
import cv2
from google.colab import files
import matplotlib.pyplot as plt

# Step 1: Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Step 2: Read image
img = cv2.imread(filename)

# Step 3: Color conversions
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# For visualization, convert HSV back to RGB
img_hsv_vis = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

# Step 4: Display all in one figure using matplotlib
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

# Original converted to RGB for display
axes[0].imshow(img_rgb)
axes[0].set_title("RGB")
axes[0].axis('off')

# Grayscale
axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title("Grayscale")
axes[1].axis('off')

# HSV visualization
axes[2].imshow(img_hsv_vis)
axes[2].set_title("HSV")
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
from google.colab import files
import matplotlib.pyplot as plt

# Step 1: Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Step 2: Read image
img = cv2.imread(filename)

# Step 3: Split channels
b, g, r = cv2.split(img)

# Convert single-channel images to 3-channel for visualization
b_disp = cv2.merge([b, b, b])
g_disp = cv2.merge([g, g, g])
r_disp = cv2.merge([r, r, r])

# Merge channels back
merged_img = cv2.merge([b, g, r])

# Step 4: Display in a single row using matplotlib
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

axes[0].imshow(cv2.cvtColor(b_disp, cv2.COLOR_BGR2RGB))
axes[0].set_title("Blue Channel")
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(g_disp, cv2.COLOR_BGR2RGB))
axes[1].set_title("Green Channel")
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(r_disp, cv2.COLOR_BGR2RGB))
axes[2].set_title("Red Channel")
axes[2].axis('off')

axes[3].imshow(cv2.cvtColor(merged_img, cv2.COLOR_BGR2RGB))
axes[3].set_title("Merged Image")
axes[3].axis('off')

plt.tight_layout()
plt.show()
```

```python
height, width = img.shape[:2]

# Define ROI (center 50% area)
start_row, start_col = int(height*0.25), int(width*0.25)
end_row, end_col = int(height*0.75), int(width*0.75)
roi = img[start_row:end_row, start_col:end_col]

print("Region of Interest (center 50%):")
cv2_imshow(roi)
```

```python
# Resize while keeping aspect ratio (half size)
half_size = (width // 2, height // 2)
resized_half = cv2.resize(img, half_size)
print("Resized to half size (aspect ratio preserved):")
cv2_imshow(resized_half)

# Resize to fixed size (changes aspect ratio)
fixed_size = (300, 150)  # width=300, height=150
resized_fixed = cv2.resize(img, fixed_size)
print("Resized to 300x150 (aspect ratio changed):")
cv2_imshow(resized_fixed)
```
