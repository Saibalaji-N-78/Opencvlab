```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img = cv2.imread(filename)

# Resize using scaling factors
scale_x, scale_y = 0.5, 0.5  # 50% of original size
resized = cv2.resize(img, (0,0), fx=scale_x, fy=scale_y)

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title(f"Resized (fx={scale_x}, fy={scale_y})")
plt.axis('off')
plt.show()
```

```python
(h, w) = img.shape[:2]
center = (w//2, h//2)
angle = 45  # degrees
scale = 1

# Rotation matrix
M = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(img, M, (w, h))

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title(f"Rotated {angle}°")
plt.axis('off')
plt.show()
```

```python
# Crop: select top-left quarter
crop = img[0:h//2, 0:w//2]

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.title("Cropped Portion")
plt.axis('off')
plt.show()
```

```python
flip_h = cv2.flip(img, 1)  # Horizontal
flip_v = cv2.flip(img, 0)  # Vertical

# Display
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(flip_h, cv2.COLOR_BGR2RGB))
plt.title("Horizontal Flip")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(flip_v, cv2.COLOR_BGR2RGB))
plt.title("Vertical Flip")
plt.axis('off')

plt.show()
```

```python
# alpha = contrast, beta = brightness
alpha = 1.5  # contrast (1.0 = original)
beta = 50    # brightness (0 = original)

adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.title(f"Brightness+Contrast (α={alpha}, β={beta})")
plt.axis('off')
plt.show()
```
