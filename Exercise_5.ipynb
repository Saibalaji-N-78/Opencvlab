```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload an image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Step 2: Read the input image
img = cv2.imread(filename)

# Step 3: Apply different filters

# 1️⃣ Averaging filter (kernel size 5x5)
avg_filtered = cv2.blur(img, (5,5))

# 2️⃣ Gaussian filter (kernel size 5x5, sigmaX=0)
gaussian_filtered = cv2.GaussianBlur(img, (5,5), 0)

# 3️⃣ Median filter (kernel size 5)
median_filtered = cv2.medianBlur(img, 5)

# 4️⃣ Bilateral filter (diameter=9, sigmaColor=75, sigmaSpace=75)
bilateral_filtered = cv2.bilateralFilter(img, 9, 75, 75)

# Step 4: Display all filtered images
fig, axes = plt.subplots(1,5, figsize=(20,5))

# Original
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original")
axes[0].axis('off')

# Averaging
axes[1].imshow(cv2.cvtColor(avg_filtered, cv2.COLOR_BGR2RGB))
axes[1].set_title("Averaging Filter")
axes[1].axis('off')

# Gaussian
axes[2].imshow(cv2.cvtColor(gaussian_filtered, cv2.COLOR_BGR2RGB))
axes[2].set_title("Gaussian Filter")
axes[2].axis('off')

# Median
axes[3].imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
axes[3].set_title("Median Filter")
axes[3].axis('off')

# Bilateral
axes[4].imshow(cv2.cvtColor(bilateral_filtered, cv2.COLOR_BGR2RGB))
axes[4].set_title("Bilateral Filter")
axes[4].axis('off')

plt.tight_layout()
plt.show()
```
