```python
# Import required libraries
import cv2                            # OpenCV library
import numpy as np                    # Numerical operations
from google.colab import files        # To upload files in Colab
from matplotlib import pyplot as plt  # To display images

# Upload image
uploaded = files.upload()

# Get uploaded file name
image_path = list(uploaded.keys()) [0]

# Read image in grayscale mode
img = cv2.imread(image_path, 0)

# Apply binary thresholding
_, binary_image = cv2.threshold (img, 50, 255, cv2.THRESH_BINARY)

# Display input and output
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before Grayscale Image")
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title("After Binary Image")
plt.imshow(binary_image, cmap='gray') # Changed 'binary' to 'binary_image'
plt.show()
```
