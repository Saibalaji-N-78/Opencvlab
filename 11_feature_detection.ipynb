```python
# Install OpenCV (if needed)
!pip install opencv-python-headless

import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files
from google.colab.patches import cv2_imshow

# Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Read image
img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# -----------------------------
# 1. Harris Corner Detection
# -----------------------------
gray_float = np.float32(gray)
harris = cv2.cornerHarris(gray_float, 2, 3, 0.04)

# Dilate for marking corners
harris = cv2.dilate(harris, None)

harris_img = img.copy()
harris_img[harris > 0.01 * harris.max()] = [0, 0, 255]

# -----------------------------
# 2. ORB Feature Detection
# -----------------------------
orb = cv2.ORB_create()
kp_orb, des_orb = orb.detectAndCompute(gray, None)

orb_img = cv2.drawKeypoints(img, kp_orb, None, color=(0,255,0), flags=0)

# -----------------------------
# 3. FAST Feature Detection
# -----------------------------
fast = cv2.FastFeatureDetector_create()
kp_fast = fast.detect(gray, None)

fast_img = cv2.drawKeypoints(img, kp_fast, None, color=(255,0,0))

# -----------------------------
# Display Results
# -----------------------------
print("Original Image")
cv2_imshow(img)

print("Harris Corner Detection")
cv2_imshow(harris_img)

print("ORB Feature Detection")
cv2_imshow(orb_img)

print("FAST Feature Detection")
cv2_imshow(fast_img)
```
