```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img = cv2.imread(filename).copy()

# Draw a blue line from top-left to bottom-right
cv2.line(img, (0, 0), (img.shape[1], img.shape[0]), color=(255,0,0), thickness=5)

# Display
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Line Drawn")
plt.axis('off')
plt.show()
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img2 = cv2.imread(filename).copy()

cv2.rectangle(img2, (50,50), (200,200), color=(0,255,0), thickness=3)
cv2.circle(img2, (300, 300), 75, color=(0,0,255), thickness=5)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title("Rectangle & Circle")
plt.axis('off')
plt.show()
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img3 = cv2.imread(filename).copy()

cv2.line(img3, (0,0), (img3.shape[1], img3.shape[0]), (255,0,0), 3)
cv2.rectangle(img3, (50,50), (200,200), (0,255,0), 3)
cv2.circle(img3, (300,300), 50, (0,0,255), -1)
cv2.ellipse(img3, (400,100), (50,25), 0, 0, 360, (255,255,0), -1)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.title("Multiple Shapes")
plt.axis('off')
plt.show()
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img4 = cv2.imread(filename).copy()

cv2.putText(img4, "OpenCV Drawing", (50,100), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,0,255), 3, cv2.LINE_AA)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
plt.title("Text on Image")
plt.axis('off')
plt.show()
```

```python
canvas = np.zeros((500,500,3), dtype=np.uint8)

cv2.line(canvas, (0,0), (500,500), (255,255,0), 5)
cv2.rectangle(canvas, (50,50), (300,300), (0,255,0), 3)
cv2.circle(canvas, (400,400), 50, (0,0,255), -1)
cv2.putText(canvas, "Canvas", (150,450), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.title("Drawing Canvas")
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img = cv2.imread(filename).copy()

# Step 2: Ask user for custom text
user_text = input("Enter the text you want to write on the image: ")

# Step 3: Write the text on the image
cv2.putText(img, user_text, (50, 100), cv2.FONT_HERSHEY_SIMPLEX,
            1.5, (255, 0, 255), 3, cv2.LINE_AA)

# Step 4: Display the image
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Text on Image")
plt.axis('off')
plt.show()
```
